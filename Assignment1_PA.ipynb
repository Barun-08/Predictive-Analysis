```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = sns.load_dataset("flights")

df.head()
```

```python
df.isnull().sum()
```

```python
df["date"] = pd.to_datetime(df["year"].astype(str) + "-" + df["month"].astype(str))

df.set_index("date", inplace=True)

ts = df["passengers"]
```

```python
import pandas as pd

# Assuming your dataframe name is df

# Convert month names to numbers
df['month_num'] = pd.to_datetime(df['month'], format='%b').dt.month

# Create a proper date column
df['date'] = pd.to_datetime(
    df['year'].astype(str) + '-' + df['month_num'].astype(str) + '-01'
)

# Set date as index
df.set_index('date', inplace=True)

# Drop helper column
df.drop('month_num', axis=1, inplace=True)

# Final output
print(df.head())

df['passengers'].plot(figsize=(12,5))

df.resample('YE')['passengers'].mean().plot()
```

```python
#simple moving avg

sma = ts.rolling(12).mean()
```

```python
sma
```

```python
window = 12
sma = []

for i in range(len(ts)):
    if i < window-1:
        sma.append(None)
    else:
        avg = sum(ts[i-window+1:i+1]) / window
        sma.append(avg)
```

```python
sma
```

```python
#cumulative moving avg
cma = ts.expanding().mean()
```

```python
cma
```

```python
cma = []
total = 0

for i in range(len(ts)):
    total += ts[i]
    avg = total / (i+1)
    cma.append(avg)
```

```python
cma
```

```python
# exponential weighted moving avg
ewma = ts.ewm(span=12, adjust=False).mean()
```

```python
ewma
```

```python
span = 12
alpha = 2/(span+1)

ewma = []
prev = ts.iloc[0]

for value in ts:
    current = alpha * value + (1-alpha) * prev
    ewma.append(current)
    prev = current
```

```python
ewma
```

```python
plt.figure(figsize=(10,5))

plt.plot(ts.values, label="Original Data")
plt.plot(sma, label="Simple Moving Average")
plt.plot(cma, label="Cumulative Moving Average")
plt.plot(ewma, label="Exponential Weighted MA")

plt.title("Time Series Analysis")
plt.xlabel("Time")
plt.ylabel("Passengers")
plt.legend()
plt.grid()

plt.show()
```
